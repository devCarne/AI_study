# 단순 선형회귀 분석

```python
import numpy as np
import matplotlib.pyplot as plt

data_x = np.arange(1,11)
data_y = 3 * np.arange(1,11) + 3 * np.random.randn(10)

plt.scatter(data_x, data_y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # 학습
lr.fit(data_x.reshape(10,1), data_y) # 비교
lr.score(data_x.reshape(10,1), data_y) # 결정계수 (R^2-score)
```

```python
print(lr.coef_) # 일치도
print(lr.intercept_)
```

```python
plt.scatter(data_x, data_y)
plt.plot(data_x, lr.predict(data_x.reshape(10,1)), color="red") #예측치
plt.show()
```

```python
import pandas as pd

tips = pd.read_csv("../contents/tips.csv")

x = tips[["total_bill"]]
y = tips["tip"]

lr = LinearRegression()
print(lr.score(x, y))
print(lr.coef_)
```
