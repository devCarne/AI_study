```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

x=np.arange(1, 11)
y=3 * x

plt.plot(x, y)
```

```python
plt.plot(y)  # index가 x축 역할
```

```python
fig = plt.figure(figsize=(4, 3))
ax1 = fig.add_subplot(2, 2, 1)  #행 2개, 열 2개, 첫 번째
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
ax1.hist(np.random.randn(100), bins=30, color="g", alpha=0.6)  #rand'n' = normal distribution, 정규분포 / bins = 가로축 갯수
ax2.scatter(np.arange(1, 31), np.arange(1, 31) + 10 * np.random.randn(30))
ax3.plot(np.random.randn(30).cumsum(), "y*--")  # y(노랑), *(별모양), --(선 모양)
ax4.plot(np.random.randn(50).cumsum(), "k--")  # cumsum = cumulative sum, 누적합 / k-- : k(블랙 색상), --(선 모양)
```

```python
fig=plt.figure(figsize=(5,2))
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), label="First", color="red")
ax.plot(np.random.randn(1000).cumsum(), label="Second", color="green")
ax.plot(np.random.randn(1000).cumsum(), label="Third", color="magenta")
ax.legend(loc="best", fontsize="small")

ticks=ax.set_xticks([0,250,500,750,1000])
labels=ax.set_xticklabels(["one", "two", "three", "four", "five"], fontsize="small", rotation=15)

ax.set_title("My Matplotlib Plot")
ax.set_xlabel("Stage")
ax.set_ylabel("level")

fig.subplots_adjust(bottom=0.5) # 서브 플롯 간격 설정

fig.savefig("matplotlib.png", dpi=400)
```
